```python
#combine all filters
df2 = df[(df['Event distance/length'].isin(['50Km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='TPE')]
df2.head(5)
```

```python
#Remove (TPE) from 
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2
```

```python
#calculate athlete age
df2['athletes_age']= 2020 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
#drop columns
df2= df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
df2
```

```python
#fix data types
df2['athletes_age'] = df2['athletes_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#rename columns
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athletes_age                   int32
```

```python
df2 = df2.rename(columns = {
'Year of event':'race_year',
'Event dates':'race-date',                   
'Event name':'race-name',              
'Event distance/length':'race_distance',         
'Event number of finishers':'no_of_finishers',      
'Athlete performance':'athlete_performance',           
'Athlete gender':'gender',                
'Athlete average speed':'average_speed',        
'Athlete ID':'id',                   
'athletes_age':'age' })
```
